```python
 # Happines Category - import four tables based in the categories in whoich we split our analysis 
import pandas as pd
df = pd.read_csv('/Users/dmb/Desktop/IRONHACK/week3/weekley_project/new tables/hapiness_category.csv')
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns

def plot_happiness_by_category(df):
    plt.figure(figsize=(12, 10))
    sns.barplot(
        data=df.sort_values("avg_score", ascending=False),
        x="avg_score", y="country_name", hue="happiness_category",
        palette={"Very happy": "green", "Moderately happy": "orange", "Less happy": "red"}
    )
    plt.title("Average Happiness Score by Country")
    plt.xlabel("Average Score")
    plt.ylabel("Country")
    plt.legend(title="Happiness Category")
    plt.tight_layout()
    plt.show()
```

```python
data = {
    "country_name": ["Switzerland", "Iceland", "Denmark", "Norway", "Canada", 
                     "Rwanda", "Benin", "Syria", "Burundi", "Togo"],
    "avg_score": [7.55, 7.53, 7.51, 7.50, 7.46, 3.47, 3.34, 3.01, 2.91, 2.84],
    "happiness_category": ["Very happy"]*5 + ["Less happy"]*5
}

df = pd.DataFrame(data)
```

```python
plot_happiness_by_category(df)
```

```python
# Happines score table 
import pandas as pd
df = pd.read_csv('/Users/dmb/Desktop/IRONHACK/week3/weekley_project/new tables/max_min_hapiness-score.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_happiness_range(df):
    df_melted = df.melt(id_vars=["country_name"], value_vars=["max_happiness", "min_happiness"],
                        var_name="Happiness Type", value_name="Happiness Score")

    plt.figure(figsize=(14, 8))
    sns.barplot(data=df_melted, x="country_name", y="Happiness Score", hue="Happiness Type")
    plt.xticks(rotation=90)
    plt.title("Max vs Min Happiness Scores by Country")
    plt.xlabel("Country")
    plt.ylabel("Happiness Score")
    plt.tight_layout()
    plt.show()
```

```python
data = {
    "country_name": ["Switzerland", "Iceland", "Denmark", "Norway", "Canada", 
                     "Rwanda", "Benin", "Syria", "Burundi", "Togo"],
    "max_happiness": [7.587, 7.561, 7.527, 7.522, 7.469, 3.465, 3.340, 3.006, 2.905, 2.839],
    "min_happiness": [7.537, 7.522, 7.504, 7.494, 7.427, 3.465, 3.340, 3.006, 2.905, 2.839]
}

df = pd.DataFrame(data)
```

```python
plot_happiness_range(df)
```
